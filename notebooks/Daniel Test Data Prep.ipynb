```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# reading with index
df = pd.read_csv("../raw_data/real_estate_sample..csv",index_col="Unnamed: 0")
```

```python
df.info
```

```python
# new data frame with split value columns
new = df["neighborhood"].str.split(" en ", n = 1, expand = True)
  
# making separate housetype
df["nhousetype"]= new[0]
  
# making separate dirty neighborhood
df["ndirty"]= new[1]
  
# cleaning neighborhood
new = df["ndirty"].str.split(" en ", n = 1, expand = True)
df["neighborhood"] = new[1] 
df = df.drop(columns="ndirty")
  
# df display
df
```

```python
df["nhousetype"].unique()
```

```python
# surface
# dropping nulls
df_nonull = df[df['surface'].notna()]
# cleaning string
new = df_nonull["surface"].str.split(" ", n = 1, expand = True)
# to float
df_nonull["surface"] = new[0].astype(float)

df_nonull
```

```python
# price
# cleaning string
new = df_nonull["price"].str.split(" ", n = 1, expand = True)
new[0] = new[0].str.replace(".","")
df_nonull["price"] = new[0]
# number filter
df_cleanprice = df_nonull[df_nonull['price'].str.isnumeric()]
# to float
df_cleanprice["price"] = df_cleanprice["price"].astype(float)
df_cleanprice
```

```python
# rooms
# cleaning string
new = df_cleanprice["rooms"].str.split(" ", n = 1, expand = True)
# to float
new[0] = new[0].astype(float)
df_cleanprice["rooms"] = new[0]
df_clean = df_cleanprice
df_clean
```
